# ETL Listing Scraper

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib as plt
```

### Craigslist

```python
# URL of page to be scraped
craigslist_url = 'https://sfbay.craigslist.org/search/vga?query=playstation+5'
# url = 'https://losangeles.craigslist.org/search/vga?query=playstation+5'
# craigslist_url = 'https://sfbay.craigslist.org/d/for-sale/search/sss?query=GeForce%20RTX%203080&sort=rel'
```

```python
response = requests.get(craigslist_url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("li", class_="result-row")
len(results)
```

```python
craigslist_df = pd.DataFrame(columns= ['title', 'price', 'link',])
craigslist_df
```

```python
for result in results:
    try:
        title = result.find('a', class_="result-title").text
        price = float(result.find('span', class_="result-price").text.replace("$","").replace(",",""))
        link = result.a['href']
        
        if (title and price and link):
            print("--------------------------------")
            print(title)
            print(price)
            print(link)
            temp_df = pd.DataFrame({'title': title, 'price': price, 'link': link}, index = [0])
            
            craigslist_df = craigslist_df.append(temp_df, ignore_index=True, sort=False)
            
        
    except AttributeError as e:
        print(e)
```

```python
craigslist_df = craigslist_df[craigslist_df['price'] > 400]
craigslist_df.sort_values(by=['price'], ascending=False)
```

```python
craigslist_df.hist()
```

### eBay

```python
ebay_url = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw=playstation+5&_sacat=139971&_ipg=200'
```

```python
response = requests.get(ebay_url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('span', class_="s-item__price")
len(results)
```

```python
for result in results:
    price = result.find('span', class_="s-item__price").text
    print("-------------------------------")
    print(price)
```
